```python
import urllib.request

url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet"
out = "green_tripdata_2025-11.parquet"
urllib.request.urlretrieve(url, out)
```

```python
url = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"
out = "taxi_zone_lookup.csv"
urllib.request.urlretrieve(url, out)
```

```python
import os

print(os.getcwd())
print(os.path.getsize("green_tripdata_2025-11.parquet"))
```

```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5433/ny_taxi")

# Parquet -> Postgres
df = pd.read_parquet("green_tripdata_2025-11.parquet")
df.to_sql(
    "green_tripdata_2025_11",
    engine,
    if_exists="replace",
    index=False,
    chunksize=10000,
    method="multi",
)

# CSV -> Postgres
zones = pd.read_csv("taxi_zone_lookup.csv")
zones.to_sql("taxi_zone_lookup", engine, if_exists="replace", index=False)
```

```python
# check if loaded correctly

engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5433/ny_taxi")

print(pd.read_sql("select count(*) as n from green_tripdata_2025_11", engine))
print(pd.read_sql("select count(*) as n from taxi_zone_lookup", engine))
```
